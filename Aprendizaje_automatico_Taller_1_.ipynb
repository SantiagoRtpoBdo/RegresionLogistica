16. Realice el entrenamiento de un algoritmo de regresiÃ³n logÃ­stica utilizando sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Seleccionar las caracterÃ­sticas con mejor correlaciÃ³n y la columna objetivo 'class_numeric'
features = ['same_srv_rate', 'dst_host_srv_count', 'dst_host_same_srv_rate', 'logged_in',
            'dst_host_srv_serror_rate', 'dst_host_serror_rate', 'serror_rate', 'srv_serror_rate', 'count']
target = 'class_numeric'

# Dividir los datos en conjuntos de entrenamiento, validaciÃ³n y pruebas
X_train = train_df[features]
y_train = train_df[target]

X_validation = validation_df[features]
y_validation = validation_df[target]

X_test = test_df[features]
y_test = test_df[target]

# Crear y entrenar un modelo de regresiÃ³n logÃ­stica
modelo = LogisticRegression(max_iter=1000)
modelo.fit(X_train, y_train)

# Realizar la predicciÃ³n en el conjunto de validaciÃ³n
y_pred_validation = modelo.predict(X_validation)

# Calcular y mostrar la precisiÃ³n del modelo en el conjunto de validaciÃ³n
precision_validation = accuracy_score(y_validation, y_pred_validation)
print("PrecisiÃ³n del modelo en el conjunto de validaciÃ³n:", precision_validation)

# Realizar la predicciÃ³n en el conjunto de pruebas
y_pred_test = modelo.predict(X_test)

# Calcular y mostrar la precisiÃ³n del modelo en el conjunto de pruebas
precision_test = accuracy_score(y_test, y_pred_test)
print("PrecisiÃ³n del modelo en el conjunto de pruebas:", precision_test)
```

17. Realice la predicciÃ³n del modelo generado utilizando el conjunto de datos de validaciÃ³n

```python
# Realizar la predicciÃ³n en el conjunto de validaciÃ³n
y_pred_validation = modelo.predict(X_validation)

# Calcular y mostrar la precisiÃ³n del modelo en el conjunto de validaciÃ³n
precision_validation = accuracy_score(y_validation, y_pred_validation)
print("PrecisiÃ³n del modelo en el conjunto de validaciÃ³n:", precision_validation)
```

18. Grafique la matriz de confusiÃ³n

```python
from sklearn.metrics import confusion_matrix

# Calcular la matriz de confusiÃ³n
confusion = confusion_matrix(y_test, y_pred_test)

# Graficar la matriz de confusiÃ³n
plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Matriz de ConfusiÃ³n')
plt.show()
```

19. Calcule la precisiÃ³n, recall, f1 score y explique los resultados obtenidos

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

# Calcular precisiÃ³n, recall y F1-score
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)

print("PrecisiÃ³n:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

20. Grafique la curva ROC y PR del modelo

```python
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, auc

# Obtener las probabilidades de la clase positiva (1) en el conjunto de pruebas
y_probs = modelo.predict_proba(X_test)[:, 1]

# Calcular la curva ROC y el Ã¡rea bajo la curva ROC (AUC-ROC)
fpr, tpr, thresholds_roc = roc_curve(y_test, y_probs)
roc_auc = roc_auc_score(y_test, y_probs)

# Calcular la curva PR y el Ã¡rea bajo la curva PR (AUC-PR)
precision, recall, thresholds_pr = precision_recall_curve(y_test, y_probs)
pr_auc = auc(recall, precision)

# Graficar la curva ROC
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Curva ROC (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')

# Graficar la curva PR
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color='darkorange', lw=2, label='Curva PR (AUC = %0.2f)' % pr_auc)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Curva Precision-Recall')
plt.legend(loc='lower left')

plt.tight_layout()
plt.show()
```
